```python
ques_id_map = {}
for qid,ques in zip(trdf.qid1, trdf.question1):
    ques_id_map[qid] = ques
    
for qid,ques in zip(trdf.qid2, trdf.question2):
    ques_id_map[qid] = ques
```

```python
max(ques_id_map.keys())
```

```python
with open("id_ques_map.json", 'w') as fh:
    json.dump(ques_id_map, fp=fh, indent=4)
```

```python
for k,v in ques_id_map.items():
    print(k,v)
    
```

```python
pair_qid_tup = list(zip(trdf.qid1,trdf.qid2))
```

```python
trdf['pair_qid_tup'] = pair_qid_tup
trdf.head()
```

```python
train_pairs, valid_pairs, y_train, y_valid = train_test_split(trdf.pair_qid_tup, trdf.is_duplicate, test_size=0.30, random_state=42)
```

```python
with open('train_pairs.txt','w') as fh:
    for i in train_pairs:
        print(i,file=fh)
        
with open('valid_pairs.txt','w') as fh:
    for i in valid_pairs:
        print(i,file=fh)
```

```python
np.savetxt("y_train.txt", X=y_train,fmt='%d')
np.savetxt("y_valid.txt", X=y_valid,fmt='%d')
```
